```python

# telecom_rdd_assignment.py
# PySpark RDD-only solution for Telecom Call Records Analysis
# -----------------------------------------------------------
# Tasks covered: 1 through 13
# Constraints respected: RDDs only (no DataFrames/Spark SQL/Pandas)

from pyspark import SparkConf, SparkContext
import os

# -----------------------------------------------------------
# Optional: Create sample CSV if not present (from the PDF)
# Comment out if you already have call_records.csv
# -----------------------------------------------------------
SAMPLE_DATA = """call_id,caller,receiver,city,call_type,duration_seconds,cost
C001,Amit,Rahul,Hyderabad,Local,180,2.5
C002,Neha,Arjun,Bangalore,STD,320,6.0
C003,Rahul,Pooja,Delhi,Local,60,1.0
C004,Pooja,Neha,Mumbai,ISD,900,25.0
C005,Arjun,Amit,Chennai,STD,400,7.5
C006,Sneha,Karan,Hyderabad,Local,240,3.0
C007,Karan,Sneha,Delhi,Local,120,2.0
C008,Riya,Vikas,Bangalore,STD,360,6.5
C009,Vikas,Riya,Mumbai,ISD,1100,30.0
C010,Anjali,Sanjay,Chennai,Local,90,1.5
C011,Farhan,Ayesha,Delhi,STD,420,7.0
C012,Ayesha,Farhan,Hyderabad,ISD,950,28.0
C013,Suresh,Divya,Bangalore,Local,150,2.0
C014,Divya,Suresh,Mumbai,STD,380,6.8
C015,Nikhil,Priya,Delhi,Local,200,2.8
C016,Priya,Nikhil,Chennai,STD,410,7.2
C017,Rohit,Kavya,Hyderabad,Local,170,2.3
C018,Kavya,Rohit,Bangalore,Local,140,2.1
C019,Manish,Tina,Mumbai,ISD,1000,27.0
C020,Tina,Manish,Delhi,STD,350,6.2
"""
CSV_PATH = "call_records.csv"
if not os.path.exists(CSV_PATH):
    with open(CSV_PATH, "w", encoding="utf-8") as f:
        f.write(SAMPLE_DATA)

# -----------------------------------------------------------
# Spark Context setup
# -----------------------------------------------------------
conf = SparkConf().setAppName("TelecomRDDAssignment").setMaster("local[*]")
sc = SparkContext(conf=conf)
sc.setLogLevel("WARN")

# -----------------------------------------------------------
# Helpers
# -----------------------------------------------------------
def parse_record(line):
    """
    Parse a CSV line (without header) into a typed tuple:
    (call_id, caller, receiver, city, call_type, duration_seconds:int, cost:float)
    Return None for malformed lines so we can filter them out safely.
    """
    parts = [p.strip() for p in line.split(",")]
    if len(parts) != 7:
        return None
    call_id, caller, receiver, city, call_type, duration_str, cost_str = parts
    try:
        duration = int(duration_str)
        cost = float(cost_str)
        return (call_id, caller, receiver, city, call_type, duration, cost)
    except:
        return None

def pretty_print(title, rows, limit=50):
    print("\n" + "="*80)
    print(title)
    print("="*80)
    for i, r in enumerate(rows):
        if i >= limit:
            print(f"... (showing {limit} of {len(rows)} rows)")
            break
        print(r)

# -----------------------------------------------------------
# Task 1: Read the CSV using textFile and display first 5 records
# -----------------------------------------------------------
lines = sc.textFile(CSV_PATH)
first5 = lines.take(5)
pretty_print("Task 1: First 5 raw lines", first5)

# -----------------------------------------------------------
# Task 2: Remove header row -> clean RDD of data lines only
# Robust approach using zipWithIndex to drop first line globally.
# -----------------------------------------------------------
header = lines.first()
data_lines = lines.filter(lambda l: l != header and l.strip() != "")
# Alternative robust method:
# data_lines = lines.zipWithIndex().filter(lambda kv: kv[1] > 0).map(lambda kv: kv[0])

# -----------------------------------------------------------
# Task 3: Split each row and convert types -> parsed RDD
# -----------------------------------------------------------
parsed = data_lines.map(parse_record).filter(lambda r: r is not None)
# Sanity check: show a few parsed records
pretty_print("Task 3: Parsed records (typed tuples)", parsed.take(5))

# Field indices for readability
IDX_CALL_ID     = 0
IDX_CALLER      = 1
IDX_RECEIVER    = 2
IDX_CITY        = 3
IDX_CALL_TYPE   = 4
IDX_DURATION    = 5
IDX_COST        = 6

# -----------------------------------------------------------
# Task 4: Total call cost per city
# -----------------------------------------------------------
total_cost_per_city = (
    parsed
    .map(lambda r: (r[IDX_CITY], r[IDX_COST]))
    .reduceByKey(lambda a, b: a + b)
)
pretty_print("Task 4: Total call cost per city", total_cost_per_city.sortBy(lambda kv: kv[0]).collect())

# -----------------------------------------------------------
# Task 5: City with the highest total call cost
# -----------------------------------------------------------
# Option A: derive from total_cost_per_city
most_expensive_city = total_cost_per_city.reduce(lambda a, b: a if a[1] >= b[1] else b)
pretty_print("Task 5: City with highest total call cost", [most_expensive_city])

# -----------------------------------------------------------
# Task 6: Total call duration per call type (Local, STD, ISD)
# -----------------------------------------------------------
total_duration_by_type = (
    parsed
    .map(lambda r: (r[IDX_CALL_TYPE], r[IDX_DURATION]))
    .reduceByKey(lambda a, b: a + b)
)
pretty_print("Task 6: Total call duration by call type", total_duration_by_type.sortBy(lambda kv: kv[0]).collect())

# -----------------------------------------------------------
# Task 7: Number of calls per city
# -----------------------------------------------------------
calls_per_city = (
    parsed
    .map(lambda r: (r[IDX_CITY], 1))
    .reduceByKey(lambda a, b: a + b)
)
pretty_print("Task 7: Number of calls per city", calls_per_city.sortBy(lambda kv: kv[0]).collect())

# -----------------------------------------------------------
# Task 8: Average call cost per city (RDD-only)
# Approach: reduceByKey over (sum_cost, count) then divide
# -----------------------------------------------------------
sum_count_per_city = (
    parsed
    .map(lambda r: (r[IDX_CITY], (r[IDX_COST], 1)))
    .reduceByKey(lambda a, b: (a[0] + b[0], a[1] + b[1]))
)
avg_cost_per_city = sum_count_per_city.mapValues(lambda sc: sc[0] / sc[1])
pretty_print("Task 8: Average call cost per city", avg_cost_per_city.sortBy(lambda kv: kv[0]).collect())

# -----------------------------------------------------------
# Task 9: High-value calls (cost > 20)
# -----------------------------------------------------------
high_value_calls = parsed.filter(lambda r: r[IDX_COST] > 20.0)
# Show selected fields for readability
high_value_selected = high_value_calls.map(lambda r: (r[IDX_CALL_ID], r[IDX_CALLER], r[IDX_RECEIVER], r[IDX_CITY], r[IDX_COST]))
pretty_print("Task 9: High-value calls (cost > 20)", high_value_selected.collect())

# -----------------------------------------------------------
# Task 10: Count ISD calls per city
# -----------------------------------------------------------
isd_calls_per_city = (
    parsed
    .filter(lambda r: r[IDX_CALL_TYPE].upper() == "ISD")
    .map(lambda r: (r[IDX_CITY], 1))
    .reduceByKey(lambda a, b: a + b)
)
pretty_print("Task 10: ISD calls per city", isd_calls_per_city.sortBy(lambda kv: kv[0]).collect())

# -----------------------------------------------------------
# Task 11: Longest call based on duration
# -----------------------------------------------------------
# Reduce by comparing duration
longest_call = parsed.reduce(lambda a, b: a if a[IDX_DURATION] >= b[IDX_DURATION] else b)
pretty_print("Task 11: Longest call (full record)", [longest_call])

# -----------------------------------------------------------
# Task 12: Total revenue generated by each caller
# -----------------------------------------------------------
revenue_per_caller = (
    parsed
    .map(lambda r: (r[IDX_CALLER], r[IDX_COST]))
    .reduceByKey(lambda a, b: a + b)
)
pretty_print("Task 12: Total revenue per caller", revenue_per_caller.sortBy(lambda kv: kv[0]).collect())

# -----------------------------------------------------------
# Task 13: Suspicious calls (duration > 900 seconds AND cost > 25)
# -----------------------------------------------------------
suspicious_calls = parsed.filter(lambda r: (r[IDX_DURATION] > 900) and (r[IDX_COST] > 25.0))
suspicious_selected = suspicious_calls.map(lambda r: (r[IDX_CALL_ID], r[IDX_CALLER], r[IDX_RECEIVER], r[IDX_CITY], r[IDX_DURATION], r[IDX_COST]))
pretty_print("Task 13: Suspicious calls (duration > 900 AND cost > 25)", suspicious_selected.collect())

## -----------------------------------------------------------
# Cleanup (optional)
# -----------------------------------------------------------
```
